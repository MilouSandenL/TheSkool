## Filtrera fram anordnare och få fram statistik för anordnaren

```python
import pandas as pd
import os

datafiler = [
    "data/resultat-ansokningsomgang-2020.xlsx",
    "data/resultat-ansokningsomgang-2021.xlsx",
    "data/resultat-ansokningsomgang-2022.xlsx",
    "data/resultat-ansokningsomgang-2023.xlsx",
    "data/resultat-ansokningsomgang-2024.xlsx"
]



# Läser och slår ihop alla filer
dfs = []

for datafil in datafiler:
    df = pd.read_excel(datafil)
    df["År"] = int(datafil[-9:-5]) # Lägger till kolumn för år
    dfs.append(df)

data = pd.concat(dfs, ignore_index=True)
data.head()    
```

```python
xlsx = pd.ExcelFile("data/resultat-ansokningsomgang-2020.xlsx")
print(xlsx.sheet_names)
```

```python
dfs = []

for datafil in datafiler:
    årtal = int(datafil[-9:-5])
    
    try:
        df = pd.read_excel(datafil, sheet_name="Tabell 3")
        df["År"] = årtal
        dfs.append(df)
    except Exception as e:
        print(f"Kunde inte läsa Tabell 3 i {datafil}: {e}")

data = pd.concat(dfs, ignore_index=True)
data.head()
```

```python
data.columns
```

```python
data["Utbildningsanordnare administrativ enhet"].unique()
```

```python
anordnare = "KYH AB"
anordnar_df = data[data["Utbildningsanordnare administrativ enhet"] == anordnare]
```

```python
# Antal ansökningar per år
anordnar_df.groupby("År")["Utbildningsnamn"].count()

# Beslut: beviljade vs ej beviljade
anordnar_df["Beslut"].value_counts()

# Beviljade utbildningar per område
anordnar_df[anordnar_df["Beslut"] == "Beviljad"].groupby("Utbildningsområde")[
    "Utbildningsnamn"
].count()
```
