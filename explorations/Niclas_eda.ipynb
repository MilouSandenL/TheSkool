```python
import pandas as pd

# Ange URL till sidan med tabellen
url = "https://www.scb.se/hitta-statistik/statistik-efter-amne/utbildning-samt-forskning-inom-hogskolan/eftergymnasial-yrkesutbildning/yrkeshogskolan/pong/tabell-och-diagram/studerande-och-examinerade/nyckeltal-studerande-och-examinerade-inom-yrkeshogskolan-2007-2024/"
# Läs in alla tabeller på sidan
tables = pd.read_html(url)

# Kontrollera hur många tabeller som hittades
print(f"Antal tabeller hittade: {len(tables)}")

# Hämta första tabellen
df = tables[0]

# Visa de första raderna i tabellen
print(df.head())

# Spara tabellen som CSV
df.to_csv("nyckeltal_yrkeshogskolan.csv", index=False, encoding="utf-8-sig")
```

```python
df = pd.read_excel("data/Bok1.xlsx")
```

```python


# Sätt årtalen som kolumnnamn
år = df.iloc[0, 1:].tolist()  # hoppa index 0, som är etiketten
df = df[1:]  # ta bort årsraden

# Sätt första kolumnen som nyckeltalsnamn
df.columns = ['Nyckeltal'] + år

# Smält om till långformat
df_long = pd.melt(df, id_vars=['Nyckeltal'], var_name='År', value_name='Värde')

# Rensa år (till int) och värde (ta bort tusentalsavgränsare, konvertera till float där det går)
df_long['År'] = df_long['År'].astype(str).str.strip()
df_long['Värde'] = df_long['Värde'].astype(str).str.replace(u'\xa0', '').str.replace(' ', '')
df_long['Värde'] = pd.to_numeric(df_long['Värde'], errors='coerce')

# Visa resultat
print(df_long.head(15))

# Spara om du vill
df_long.to_csv("data/nyckeltal_yrkeshogskolan_long.csv", index=False, encoding="utf-8-sig")
```

```python
import pandas as pd

def load_trend_data():
    df = pd.read_csv("data/trender.csv", encoding="latin1")
    cols_to_drop = ['ålder', 'kön']
    for col in cols_to_drop:
        if col in df.columns:
            df = df.drop(columns=[col])
    return df
```
